## Visualization & Explanatory Analysis

### Pertanyaan 1:

  Bagaimana cuaca (weathersit) memengaruhi penggunaan sepeda?


```python
byweathersit_df = hour_df.groupby(by="weathersit").cnt.nunique().reset_index()
byweathersit_df.rename(columns={
    "cnt": "jumlah_sepeda_yang_disewa"
}, inplace=True)

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
byweathersit_df
plt.figure(figsize=(10,5))
sb.barplot(
    y="jumlah_sepeda_yang_disewa",
    x="weathersit",
    data=byweathersit_df.sort_values(by="jumlah_sepeda_yang_disewa", ascending=False),
    palette=colors
)
plt.title("Number of weathersit on hours by jumlah sepeda", loc="center", fontsize=15)
plt.xlabel("cuaca")
plt.ylabel("banyak sepeda yang disewa")
plt.show()
```

```python
byweathersit_df = day_df.groupby(by="weathersit").cnt.nunique().reset_index()
byweathersit_df.rename(columns={
    "cnt": "jumlah_sepeda_yang_disewa"
}, inplace=True)
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
byweathersit_df
plt.figure(figsize=(10,5))
sb.barplot(
    y="jumlah_sepeda_yang_disewa",
    x="weathersit",
    data=byweathersit_df.sort_values(by="jumlah_sepeda_yang_disewa", ascending=False),
    palette=colors
)
plt.title("Number of weathersit on day by jumlah sepeda", loc="center", fontsize=15)
plt.xlabel("cuaca")
plt.ylabel("banyak sepeda yang disewa")
plt.show()
```

### Pertanyaan 2:

  Bagaimana tren penggunaan sepeda per bulan pada tahun 2011? Apakah ada bulan-bulan tertentu yang memiliki peningkatan atau penurunan signifikan dalam jumlah pemesanan sepeda?

```python
monthly_order_df = hour_df.resample(rule='M', on='dteday').agg({
    "instant": "nunique",
    "cnt": "sum"
})

hour_df_2011 = monthly_order_df[(monthly_order_df.index >= '2011-01-01') & (monthly_order_df.index < '2012-01-01')]

hour_df_2011.index = hour_df_2011.index.strftime('%B')

hour_df_2011 = hour_df_2011.reset_index()
hour_df_2011.rename(columns={
    "instant": "order",
    "cnt": "count"
}, inplace=True)

plt.figure(figsize=(12,17))
plt.plot(
    hour_df_2011["dteday"],
    hour_df_2011["count"],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Number of orders per on hour (perjam) Month (2011)")
plt.xlabel(None)
plt.ylabel("Jumlah sepeda yang tersewa")
plt.show()
```

```python
monthly_order_df = day_df.resample(rule='M', on='dteday').agg({
    "instant": "nunique",
    "cnt": "sum"
})

hour_df_2011 = monthly_order_df[(monthly_order_df.index >= '2011-01-01') & (monthly_order_df.index < '2012-01-01')]

hour_df_2011.index = hour_df_2011.index.strftime('%B')

hour_df_2011 = hour_df_2011.reset_index()
hour_df_2011.rename(columns={
    "instant": "order",
    "cnt": "count"
}, inplace=True)

plt.figure(figsize=(12,17))
plt.plot(
    hour_df_2011["dteday"],
    hour_df_2011["count"],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Number of orders on day(perhari) per Month (2011)")
plt.xlabel(None)
plt.ylabel("Jumlah sepeda yang tersewa")
plt.show()
```

### Pertanyan 3

Bagaimana tren penggunaan sepeda per bulan pada tahun 2012? Apakah ada bulan-bulan tertentu yang memiliki peningkatan atau penurunan signifikan dalam jumlah pemesanan sepeda?

```python
monthly_order_df = hour_df.resample(rule='M', on='dteday').agg({
    "instant": "nunique",
    "cnt": "sum"
})

hour_df_2012 = monthly_order_df[(monthly_order_df.index >= '2012-01-01') & (monthly_order_df.index < '2013-01-01')]

hour_df_2012.index = hour_df_2012.index.strftime('%B')

hour_df_2012 = hour_df_2012.reset_index()
hour_df_2012.rename(columns={
    "instant": "order",
    "cnt": "count"
}, inplace=True)

plt.figure(figsize=(12,17))
plt.plot(
    hour_df_2012["dteday"],
    hour_df_2012["count"],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Number of orders (perjam) per Month (2012)")
plt.xlabel(None)
plt.ylabel("Jumlah sepeda yang tersewa")
plt.show()
```

```python
monthly_order_df = day_df.resample(rule='M', on='dteday').agg({
    "instant": "nunique",
    "cnt": "sum"
})

hour_df_2012 = monthly_order_df[(monthly_order_df.index >= '2012-01-01') & (monthly_order_df.index < '2013-01-01')]

hour_df_2012.index = hour_df_2012.index.strftime('%B')

hour_df_2012 = hour_df_2012.reset_index()
hour_df_2012.rename(columns={
    "instant": "order",
    "cnt": "count"
}, inplace=True)

plt.figure(figsize=(12,17))
plt.plot(
    hour_df_2012["dteday"],
    hour_df_2012["count"],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Number of orders (perhari) per Month (2012)")
plt.xlabel(None)
plt.ylabel("Jumlah sepeda yang tersewa")
plt.show()
```
